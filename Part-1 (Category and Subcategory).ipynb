### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
columns_list = crowdfunding_info_df.columns.tolist()

# Print the list of column names
print(columns_list)
```

```python
# Assign the category and subcategory values to category and subcategory columns.
# Split the 'category & sub-category' column into 'category' and 'subcategory'
split_columns = crowdfunding_info_df['category & sub-category'].str.split('/', n=1, expand=True)

# Assign the values to 'category' and 'subcategory' columns
crowdfunding_info_df['category'] = split_columns[0]

# If 'subcategory' exists in the split columns, assign it; otherwise, create an 'subcategory' column with None values
if len(split_columns.columns) > 1:
    crowdfunding_info_df['subcategory'] = split_columns[1]
else:
    crowdfunding_info_df['subcategory'] = None

# Display the updated DataFrame
crowdfunding_info_df.head()
```

```python
# Get unique categories and subcategories
unique_categories = crowdfunding_info_df['category'].unique().tolist()
unique_subcategories = crowdfunding_info_df['subcategory'].unique().tolist()

# Print the lists
print("Unique Categories:")
print(unique_categories)

print("\nUnique Subcategories:")
print(unique_subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists
num_unique_categories = len(unique_categories)
num_unique_subcategories = len(unique_subcategories)

# Print the counts
print("Number of Distinct Categories:", num_unique_categories)
print("Number of Distinct Subcategories:", num_unique_subcategories)
```

```python
import numpy as np

# Create NumPy arrays for categories (1-9) and subcategories (1-24)
categories_array = np.arange(1, 10)
subcategories_array = np.arange(1, 25)

# Print the arrays
print("Categories Array:", categories_array)
print("Subcategories Array:", subcategories_array)
```

```python
# Add "cat" to each category_id using list comprehension
categories_with_cat = [f"cat{category_id}" for category_id in categories_array]

# Add "subcat" to each subcategory_id using list comprehension
subcategories_with_subcat = [f"subcat{subcategory_id}" for subcategory_id in subcategories_array]

# Print the modified lists
print("Categories with 'cat':", categories_with_cat)
print("Subcategories with 'subcat':", subcategories_with_subcat)
```

```python
import pandas as pd

# Create a category DataFrame
category_df = pd.DataFrame({
    'category_id': [f'cat{category_id}' for category_id in categories_array],
    'category': unique_categories
})

# Create a subcategory DataFrame
subcategory_df = pd.DataFrame({
    'subcategory_id': [f'subcat{subcategory_id}' for subcategory_id in subcategories_array],
    'subcategory': unique_subcategories
})

# Print the DataFrames
print("Category DataFrame:")
print(category_df)

print("\nSubcategory DataFrame:")
print(subcategory_df)
```
